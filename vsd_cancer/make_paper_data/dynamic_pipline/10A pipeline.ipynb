```python
from catch22 import catch22_all
tc_filt=mcf_filt.iloc[:,401:]
features=pd.DataFrame([])
for i in range(tc_filt.shape[0]):
    df=tc_filt.iloc[i,:]
    #Feature Extraction
    catch22_out=catch22_all(df)
    features=features.append(catch22_out,ignore_index=True)
values=features['values'].apply(pd.Series)
values.columns=['DN_HistogramMode_5','DN_HistogramMode_10','CO_f1ecac','CO_FirstMin_ac','CO_HistogramAMI_even_2_5','CO_trev_1_num','MD_hrv_classic_pnn40','SB_BinaryStats_mean_longstretch1','SB_TransitionMatrix_3ac_sumdiagcov','PD_PeriodicityWang_th0_01','CO_Embed2_Dist_tau_d_expfit_meandiff','IN_AutoMutualInfoStats_40_gaussian_fmmi','FC_LocalSimple_mean1_tauresrat','DN_OutlierInclude_p_001_mdrmd','DN_OutlierInclude_n_001_mdrmd','SP_Summaries_welch_rect_area_5_1','SB_BinaryStats_diff_longstretch0','SB_MotifThree_quantile_hh','SC_FluctAnal_2_rsrangefit_50_1_logi_prop_r1','SC_FluctAnal_2_dfa_50_1_2_logi_prop_r1','SP_Summaries_welch_rect_centroid','FC_LocalSimple_mean3_stderr']

import seaborn as sns
for i in range(values.shape[1]):
    sns.histplot(values.iloc[:,i],bins=50)
    plt.title(values.columns[i])
    #plt.show()
    
f=values
f=f.drop(['IN_AutoMutualInfoStats_40_gaussian_fmmi'],axis=1)
import scipy.stats as spstats
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler(feature_range=(0.0001,1))
fb = min_max_scaler.fit_transform(f)
fb=pd.DataFrame(fb)

for i in range(f.shape[1]):
    feature=np.array(fb.iloc[:,i])
    l,opt_lambda=spstats.boxcox(feature)
    #print('Optimal lambda value:', opt_lambda)
    f.iloc[:,i]=spstats.boxcox(fb.iloc[:,i],lmbda=opt_lambda)

fb= MinMaxScaler(feature_range=(0,1)).fit_transform(f)
fb= pd.DataFrame(fb)
fb.columns=['DN_HistogramMode_5','DN_HistogramMode_10','CO_f1ecac','CO_FirstMin_ac','CO_HistogramAMI_even_2_5',
            'CO_trev_1_num','MD_hrv_classic_pnn40','SB_BinaryStats_mean_longstretch1','SB_TransitionMatrix_3ac_sumdiagcov',
            'PD_PeriodicityWang_th0_01','CO_Embed2_Dist_tau_d_expfit_meandiff','FC_LocalSimple_mean1_tauresrat',
            'DN_OutlierInclude_p_001_mdrmd','DN_OutlierInclude_n_001_mdrmd','SP_Summaries_welch_rect_area_5_1',
            'SB_BinaryStats_diff_longstretch0','SB_MotifThree_quantile_hh','SC_FluctAnal_2_rsrangefit_50_1_logi_prop_r1',
            'SC_FluctAnal_2_dfa_50_1_2_logi_prop_r1','SP_Summaries_welch_rect_centroid','FC_LocalSimple_mean3_stderr']

for i in range(fb.shape[1]):
    sns.histplot(fb.iloc[:,i],bins=50)
    #fig = plt.figure()
    #res = stats.probplot(feature.iloc[:,i], plot=plt)
    plt.title('Scale[0.0001,1]+Boxcox')
    plt.ylabel('Number')
    plt.xlabel(fb.columns[i])
    #plt.savefig(fb.columns[i])
    #plt.show()

from sklearn.decomposition import PCA
mcf_filt.index=np.arange(mcf_filt.shape[0])
pca = PCA(n_components=0.8)
new_pca = pd.DataFrame(pca.fit_transform(fb))
print('Variance of PCA:')
print(pca.explained_variance_)

print('Variance Ratio:')
print(pca.explained_variance_ratio_) 
importance = pca.explained_variance_ratio_
c=pca.components_

#print('输出主成分系数阵:默认列是指标，行是指标在成为上的系数')
#print(pca.components_)

print('Number of principle components: {}'.format(pca.n_components_))

t0=new_pca[notes['expt']=='MCF10A']
t1=new_pca[notes['expt']=='MCF10A_TGFB']

plt.figure(dpi=300)

plt.scatter(t0.iloc[:,0],t0.iloc[:,1],marker='o',label='MCF10A')
plt.scatter(t1.iloc[:,0],t1.iloc[:,1],marker='o',label='MCF-10A + TGF-β')
plt.xlabel(f'1st PC:{round(importance[0]*100,1)}%',fontsize=14)
plt.ylabel(f'2nd PC:{round(importance[1]*100,1)}%',fontsize=14)

kwargs={'linestyle':'-', 'lw':0.1, 'width':0.02} 
for i in range(c.shape[1]):
    plt.arrow(0,0,c[0,i],c[1,i],alpha=0.5,ec='green',fc='green',**kwargs)
    plt.text(c[0,i],c[1,i],f'{i+1}', ha="center", va="center")
#plt.xlabel(f'PC1:{round(importance[0]*100,1)}%',fontsize=14)  
#plt.ylabel(f'PC2:{round(importance[1]*100,1)}%',fontsize=14)
plt.xlabel(r'PC1',fontsize=20)  
plt.ylabel(r'PC2',fontsize=20)
plt.legend(fontsize=16,frameon=False,loc=[1,0.5])
#plt.legend(fontsize=16,frameon=False,loc='upper left')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
#plt.xticks([])
#plt.yticks([])

plt.legend(fontsize=16,frameon=False,loc=[1,0.5])
#plt.title(r'PCA - MCF10A/MCF10A+TGF$\beta$',fontsize=16)

from sklearn.manifold import TSNE

tsne=TSNE()
new_tsne = pd.DataFrame(tsne.fit_transform(fb))
#notes['condition']=notes['expt']
t1=new_tsne[notes['expt']=='MCF10A']
t2=new_tsne[notes['expt']=='MCF10A_TGFB']

fig=plt.figure(dpi=300)
ax=fig.add_subplot(111)

ax.scatter(t1.iloc[:,0],t1.iloc[:,1],label='MCF-10A',marker='o')
ax.scatter(t2.iloc[:,0],t2.iloc[:,1],label='MCF-10A + TGF-β',marker='o')

#ax.set_xlim([-1.5,1.5])
#ax.set_ylim([-0.75,1.25])
plt.xlabel(r't-SNE1',fontsize=20)  
plt.ylabel(r't-SNE2',fontsize=20)
plt.legend(fontsize=16,frameon=False,loc=[1,0.5])
#plt.legend(fontsize=16,frameon=False,loc='upper left')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xticks([])
plt.yticks([])
#plt.savefig('tsne of 10a.eps',format='eps',dpi=600)
#plt.savefig('tsne of 10a.png',dpi=600)
```

```python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.decomposition import PCA



X = fb  
y = np.array([0]*27 + [1]*33)

pca = PCA(n_components=0.8)
Xreduced = pca.fit_transform(X)
Xreduced = Xreduced[:,:2]

def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out

model = svm.SVC(kernel='linear')
clf = model.fit(Xreduced, y)

fig, ax = plt.subplots(dpi=300)
# title for the plots
#title = ('Decision surface of linear SVC ')
# Set-up grid for plotting.
X0, X1 = Xreduced[:, 0], Xreduced[:, 1]
xx, yy = make_meshgrid(X0, X1)

plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, edgecolors='k')
#handles, labels = ax.get_legend_handles_labels()
#ax.legend(handles, 
          #['MCF-10A', 'MCF-10A + TGF-β'], 
        #scatterpoints=1,fontsize=16,frameon=False,loc=[1,0.5])
ax.scatter(X0[:27], X1[:27],  color='mediumslateblue', s=20, edgecolors='k',label='MCF-10A')
ax.scatter(X0[27:], X1[27:],  color='lightcoral', s=20, edgecolors='k',label='MCF-10A + TGF-β')
ax.set_ylabel('PC2',fontsize=20)
ax.set_xlabel('PC1',fontsize=20)
ax.set_xticks(())
ax.set_yticks(())
#ax.set_title('Decison surface using the PCA transformed/projected features')
ax.legend(fontsize=16,frameon=False,loc=[1,0.5])
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlim([-1.5,1.5])
plt.ylim([-1,1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
model = svm.SVC(kernel='linear')
X_train, X_test, y_train, y_test = train_test_split(Xreduced, y, test_size=0.25, random_state=33)
clf = model.fit(X_train, y_train)
#c_y_predict=model.predict(X_test)
#scores.append(model.score(X_test,y_test))
print('Accuracy is ',model.score(X_test,y_test))
```

```python
from sklearn.decomposition import PCA
mcf_filt.index=np.arange(mcf_filt.shape[0])
pca = PCA(n_components=0.8)
new_pca = pd.DataFrame(pca.fit_transform(fb))
print('Variance of PCA:')
print(pca.explained_variance_)

print('Variance Ratio:')
print(pca.explained_variance_ratio_) 
importance = pca.explained_variance_ratio_
c=pca.components_

#print('输出主成分系数阵:默认列是指标，行是指标在成为上的系数')
#print(pca.components_)

print('Number of principle components: {}'.format(pca.n_components_))

t0=new_pca[notes['expt']=='MCF10A']
t1=new_pca[notes['expt']=='MCF10A_TGFB']

plt.figure(dpi=300)

plt.scatter(t0.iloc[:,0],t0.iloc[:,1],marker='o',label='MCF10A')
plt.scatter(t1.iloc[:,0],t1.iloc[:,1],marker='o',label='MCF-10A + TGF-β',c='green')
#plt.xlabel(f'1st PC:{round(importance[0]*100,1)}%',fontsize=14)
#plt.ylabel(f'2nd PC:{round(importance[1]*100,1)}%',fontsize=14)

kwargs={'linestyle':'-', 'lw':0.1, 'width':0.02} 
#for i in range(c.shape[1]):
    #plt.arrow(0,0,c[0,i],c[1,i],alpha=0.5,ec='green',fc='green',**kwargs)
    #plt.text(c[0,i],c[1,i],f'{i+1}', ha="center", va="center")
#plt.xlabel(f'PC1:{round(importance[0]*100,1)}%',fontsize=14)  
#plt.ylabel(f'PC2:{round(importance[1]*100,1)}%',fontsize=14)
plt.xlabel(r'PC 1',fontsize=20)  
plt.ylabel(r'PC 2',fontsize=20)
plt.legend(fontsize=16,frameon=False,loc=[1,0.5])
#plt.legend(fontsize=16,frameon=False,loc='upper left')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xticks([])
plt.yticks([])

plt.legend(fontsize=16,frameon=False,loc=[1,0.5])
#plt.title(r'PCA - MCF10A/MCF10A+TGF$\beta$',fontsize=16)
#plt.scatter(clf.support_vectors_[:, 0],
            #clf.support_vectors_[:, 1],
            #s=80, edgecolors="k", facecolors="none")
Xreduced = pd.concat([t0,t1],axis=0)
#Xreduced=Xreduced.iloc[:,:2]
model = svm.SVC(kernel='linear')
clf = model.fit(Xreduced, y)
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-1.2, 1.5)
yy = a * xx - (clf.intercept_[0]) / w[1]

plt.plot(xx, yy, 'k--')
plt.xlim([-1.2,1.5])
plt.ylim([-1,1])
#plt.savefig('pca of 10As.pdf',format='pdf',bbox_inches='tight',dpi=300)

X_train, X_test, y_train, y_test = train_test_split(Xreduced, y, test_size=0.25, random_state=33)
clf = model.fit(X_train, y_train)
y_predict=model.predict(X_test)
#scores.append(model.score(X_test,y_test))
print('Accuracy is ',model.score(X_test,y_test))
```
